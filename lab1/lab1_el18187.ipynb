## Α. Κατασκευή και οπτικοποίηση σύνθετων τύπων δικτύου

### Τοπολογία REG για βαθμό πλέγματος $d = 4$

```python
random.seed(seedno)
n=47 #μέγεθος δικτύου: αντικαταστήστε το x με το τελευταίο ψηφίο του αριθμού μητρώου σας.
d=4
# για την κατασκευή του REG χρησιμοποιήστε την watts_strogatz_graph από το networkx. 
# Επιλέξτε προσεκτικά την τιμή της παραμέτρου p. 
REG = nx.watts_strogatz_graph(n, d, 0)

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(REG)
```

### Τοπολογία RG (ER) για αριθμό συνδέσεων $M = 150$

```python
random.seed(seedno)
n = 47 
M = 150

# για την κατασκευή του RGER χρησιμοποιήστε την gnm_random_graph από το networkx
RGER = nx.gnm_random_graph(n, M, seed=1)

# οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(RGER)
```

### Τοπολογία RGG  με ακτίνα σύνδεσης $R=0.35$

```python
random.seed(seedno)
n = 47
R = 0.35

# για την κατασκευή του RGG χρησιμοποιήστε την random_geometric_graph από το networkx
RGG = nx.random_geometric_graph(n, R)

#οπτικοποίηση: https://networkx.org/documentation/stable/auto_examples/drawing/plot_random_geometric_graph.html

# position is stored as node attribute data for random_geometric_graph
pos = nx.get_node_attributes(RGG, "pos")

# find node near center (0.5,0.5)
dmin = 1
ncenter = 0
for n in pos:
    x, y = pos[n]
    d = (x - 0.5) ** 2 + (y - 0.5) ** 2
    if d < dmin:
        ncenter = n
        dmin = d

# color by path length from node near center
p = dict(nx.single_source_shortest_path_length(RGG, ncenter))

plt.figure(figsize=(8, 8))
nx.draw_networkx_edges(RGG, pos, alpha=0.4)
nx.draw_networkx_nodes(
    RGG,
    pos,
    nodelist=list(p.keys()),
    node_size=80,
    node_color=list(p.values()),
    cmap=plt.cm.Reds_r,
)

plt.xlim(-0.05, 1.05)
plt.ylim(-0.05, 1.05)
plt.axis("off")
plt.show()
```

### Τοπολογία SF με παράμετρο επισύναψης $d=4$

```python
random.seed(seedno)
n = 47
d = 4

# για την κατασκευή του SF χρησιμοποιήστε την barabasi_albert_graph από το networkx
SF = nx.barabasi_albert_graph(n, d) 

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(SF)
```

### Τοπολογία SW με βαθμό αρχικού πλέγματος $d=4$ και πιθανότητα ανασύνδεσης $p=0.25$

```python
random.seed(seedno)
n = 47
d = 4
p = 0.25

# για την κατασκευή του SW χρησιμοποιήστε την watts_strogatz_graph από το networkx. 
SW = nx.watts_strogatz_graph(n, d, p) 

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(SW)
```

### B1. Συντελεστής ομαδοποίησης (Clustering coefficient)

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):
#### Β1.1 Να υπολογιστεί ο συντελεστής ομαδοποίησης κάθε κόμβου και να αναπαρασταθεί η κατανομή του.
#### Β1.2  Να υπολογιστεί ο μέσος συντελεστής ομαδοποίησης και η διασπορά του.
#### Β.1.3 Να βρεθούν οι κόμβοι με το μικρότερο και το μεγαλύτερο συντελεστή ομαδοποίησης. 
#### Να σχολιαστούν τα αποτελέσματα.

```python
import numpy as np
def cc(G, name):
    #B.1.1
    clust_coeffs = nx.clustering(G)
    values = list(clust_coeffs.values())
    plt.hist(values)
    plt.title(name + ': Clustering Coefficient Distribution')
    plt.xlabel('Clustering Coefficient Value')
    plt.ylabel('# nodes')
    plt.show()
    #B.1.2
    mean = np.mean(values)
    var = np.var(values)
    print('Average Clustering Coefficient: ' + str(mean))
    print('Clustering Coefficient Variance: ' + str(var))
    #B.1.3
    print('Node with minimum Clustering Coefficient: ' + str(min(clust_coeffs, key=clust_coeffs.get)))
    print('Node with maximum Clustering Coefficient: ' + str(max(clust_coeffs, key=clust_coeffs.get)))
```

```python
cc(REG, 'REG')
```

Όλοι οι κόμβοι έχουν τον ίδιο συντελεστή ομαδοποίησης (τιμή 0.5) και για αυτό η διασπορά είναι μηδενική. Αυτό συμβαίνει καθώς όλοι οι κόμβοι έχουν τον ίδιο βαθμό και συνδέονται μόνο με τους 4 κοντινότερους γείτονές τους.

ΕΦόσον έχουμε γραάφο με 47 κόμβους με 150 ακμές, τοποθετημένες με τυχαίο τρόπο, αυτό σημαίνει πως δεν υπάρχει ίση κατανομή βαθμού στους κόμβους. Δηλαδή, μερικοί θα συνδέονται με πολλούς, ενώ άλλοι με πολύ λιγότερους. Αυτο σημαίνει πως οι γείτονες του κάθε κόμβου μάλλον δεν θα συνδέονται μεταξύ τους. Επομένως και έχουμε μικρό average clustering coefficient.

Για το μοντέλο RGG περιμένουμε σχετικά μεγάλο average clustering coefficient εφόσον ο τρόπος κατασκευής του γράφου βασίζεται στην σύνδεση γειτόνων ενός κόμβου μέσα σε μία ορισμένη ακτίνα R. Αυτό σημαίνει πως θα έχουμε σχετικά μικρό mean average clustering coefficient. Παρατηρούμε στο διάγραμμα κατανομής πως πολλοί κόμβοι έχουν σχεδόν μηδενικό clustering coefficient.

### Β2. Μέσο μήκος μονοπατιού - Διάμετρος - Ακτίνα γράφου

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):
#### Β2.1 Nα υπολογιστεί το μέσο μήκος μονοπατιού και η διασπορά του. Προτείνεται η χρήση της συνάρτησης  all_pairs_shortest_path_length του networkx.
#### B2.2 Να υπολογιστεί η διάμετρος (συνάρτηση diameter) και η ακτίνα (συνάρτηση radius).
#### Για τα παραπάνω πραγματοποιήστε συγκρίσεις και εξηγήστε με βάση τα χαρακτηριστικά της κάθε τοπολογίας όπως παρουσιάζονται στη θεωρία.

```python
def compute_path_lengths(G):
    #B.2.1
    length = dict(nx.all_pairs_shortest_path_length(G))
    mean = nx.average_shortest_path_length(G)
    print('Average shortest path length: ' + str(mean))
    
    var = 0
    for i in range(0, len(length)):
        for j in range(i, len(length[i])):
            var += (length[i][j]-mean)**2     
    print('Variance path length: ' + str(var))
    
    print('Diameter: ' + str(nx.diameter(G)))
    print('Radius: ' + str(nx.radius(G)))
```

### Σύγκριση αποτελεσμάτων του παραπάνω πίνακα

 Παρατηρούμε για το REG πως το μέσο μήκος μονοπατιού είναι 6 και είναι σταθερό αφού όλοι οι κόμβοι έχουν το ίδιο degree. Το μήκος αυτό είναι μεγάλο σε σχέση με τα υπόλοιπα μοντέλα. Ο τρόπος κατασκευής του απλού γράφου REG δεν επιτρέπει την δημιουργία μικρών ελάχιστων μονοπατιών.

Για το RGER μοντέλο περιμένουμε σχετικά μικρό μέσο μήκος μονοπατιού. Επειδή έχουμε μεγάλο μέσο βαθμό κόμβου μειώνεται το μέσο μήκος μονοπατιού.

Και για το RGG θα περιμέναμε μεγάλο μέσο μήκος μονοπατιού. Ωστόσο, αυτό οφείλεται από την απόσταση R^d μεταξύ των κόμβων. Για τιμές του R που επιτρέπει την δημιουργία ακμών σε μεγάλες αποστάσεις αναμένεται μείωση του μέσο μήκους μονοπατιού. Στην δική μας περίπτωση, η τιμή του R είναι σχετικά μεγάλη και επομένος το μέσο μήκος μονοπατιού μικρό.

Για το SF μοντέλο περιμένουμε πολύ μικρό μήκος μέσου μονοπατιού, όπως και παρατηρούμε εφόσον η τιμή είναι η μικρότερη σε σχέση με τα άλλα μοντέλα. Επίσης, παρατηρούμε πως και η διασπορά είναι η μικρότερη σε σχέση με τα υπόλοιπα. Αυτό υποδηλώνει και την δομή του SF, δηλαδή πως οι κόμβοι είτε είναι κόμβοι που έχουν μεγάλο βαθμό είτε κόμβοι που συνδέονται με σχετικά λίγα hops σε αυτούς τους κόμβους.

Για το SW περιμένουμε να λάβουμε ένα σχετικά μικρό μέσο μήκος μονοπατιού το οποίο όμως εξαρτάται από το πλήθος των κόμβων. Εδώ η τιμή είναι 3.11. Έγινε δοκιμή για μεγαλύτερο πλήθος κόμβων και η τιμή του μέσου μήκους μονοπατιού αυξήθηκε αξιοσημείωτα.

#### B.2.3 Για κάθε τοπολογία, ποιος κόμβος συνδέεται με τους περισσότερους κόμβους με μήκος ελάχιστου μονοπατιού ίσο με τη διάμετρο της τοπολογίας?

```python
def find_nodes(G):
    length = dict(nx.all_pairs_shortest_path_length(G))
    diameter = nx.diameter(G)
    node = 0
    max_occurance = 0
    for i in range(0, len(length)):
        occurance = 0
        for j in range (0, len(length[i])):
            if (length[i][j] == diameter):
                occurance += 1
        if (max_occurance < occurance):
            node = i
            max_occurance = occurance
    return (node, max_occurance)
```

```python
# να προσδιοριστεί το label του κόμβου και το σύνολο των κόμβων 
# με τους οποίους συνδέεται με μήκος ελάχιστου μονοπατιού ΄ίσο με τη διάμετρο. 
```

```python
(label, nodes) = find_nodes(REG)
print('REG:')
print('Node label: ' + str(label))
print('Number of connected nodes: ' + str(nodes))
print('\n')
(label, nodes) = find_nodes(RGER)
print('RGER:')
print('Node label: ' + str(label))
print('Number of connected nodes: ' + str(nodes))
print('\n')
(label, nodes) = find_nodes(RGG)
print('RGG:')
print('Node label: ' + str(label))
print('Number of connected nodes: ' + str(nodes))
print('\n')
(label, nodes) = find_nodes(SF)
print('SF:')
print('Node label: ' + str(label))
print('Number of connected nodes: ' + str(nodes))
print('\n')
(label, nodes) = find_nodes(SW)
print('SW:')
print('Node label: ' + str(label))
print('Number of connected nodes: ' + str(nodes))
```

#### Περιφέρεια του γράφου ορίζεται να είναι το σύνολο των κόμβων του που έχουν εκκεντρότητα ίση με τη διάμετρό του. 
#### Κέντρο του γράφου ορίζεται να είναι το σύνολο των κόμβων του που έχουν εκκεντρότητα ίση με την ακτίνα του.  
#### Β.2.4 Για κάθε τοπολογία, να βρεθεί η περιφέρεια και το κέντρο της και να σχολιαστούν τα αποτελέσματα.

```python
def find_periphery(G):
    nodes = nx.periphery(G)
    print("Periphery Nodes:\n" + str(nodes))
    nx.draw(G.subgraph(nodes), pos=pos, node_color="red", edge_color="red")
    return
```

```python
def find_center(G):
    nodes = nx.center(G)
    print("Center Nodes:\n" + str(nodes))
    nx.draw(G.subgraph(nodes), pos=pos, node_color="green", edge_color="darkgreen")
    return
```

### Γ1. Κεντρικότητα βαθμού (Degree centrality)

```python
def find_degree(G, name):
    degrees = [d for n, d in G.degree()]
    plt.title(name + ": Node Degree Distribution")
    plt.hist(degrees)
    plt.xlabel('Node Degree')
    plt.ylabel('# Nodes')
    summ = 0
    for degree in degrees:
        summ += degree
    mean = summ/len(G.nodes())
    print("Average Degree: " + str(mean))
    
    var = 0
    var = np.var(list(degrees))
    print("Variance Degree: " + str(var))
    
    return
```

### Γ4. Kεντρικότητα Katz
##### Γ4.1 Να αναπαρασταθεί η κατανομή της κεντρικότητας Katz των κόμβων για κάθε τοπολογία. 
Για τον προσδιορισμό της παραμέτρου $alpha$ της συνάρτησης katz_centrality,  να υπολογιστεί η μεγαλύτερη ιδιοτιμή $λ_{max}$ του πίνακα γειτνίασης με τη συνάρτηση adjacency_spectrum του networkx. 
H $alpha$ ορίζεται να είναι ιση με $\frac{1}{λ_{max}}-0.01$.
##### Γ4.2 Να υπολογιστεί η μέση  κεντρικότητα Katz  καθώς και η διασπορά της για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
def find_katz(G, name):
    alpha = 1/max(nx.adjacency_spectrum(G)) - 0.01
    katz_centrality = nx.katz_centrality(G, alpha)
    values = list(katz_centrality.values())
    plt.title(name + ": Katz Centrality Distribution")
    plt.hist(values)
    plt.xlabel('Katz Centrality Value')
    plt.ylabel('# Nodes')
    mean = np.mean(values)
    var = np.var(values)
    print(name)
    print("Average Katz Centrality: " + str(mean))
    print("Variance of Katz Centrality: " + str(var))
    print('\n')
```

### Γ5. Αλγόριθμος PageRank σε πραγματικό δίκτυο
#####  Να εκτελεστεί ο αλγόριθμος pagerank στην 41η μεγαλύτερη συνδεδεμένη συνιστώσα (index=40) του κατευθυνόμενου δίκτυου web-Stanford του οποίου οι ακμές δίνονται στο αρχείο web-Standord.txt. 
##### Για να υπολογίσετε τις συνδεδεμένες συνιστώσες, χρησιμοποιήστε τη συνάρτηση weakly_connected_components του networkx.
Περισσότερες πληροφορίες για το dataset δίνονται στο link https://snap.stanford.edu/data/web-Stanford.html.
##### Να σχολιαστούν τα αποτελέσματα.

```python
G_web=nx.read_edgelist("web-Stanford.txt",create_using=nx.DiGraph())
```

```python
#find the 40th largest weakly connected component of the graph web-Stanford
weakly_components_sorted = sorted(nx.weakly_connected_components(G_web), key=len, reverse=True)
G40 = nx.subgraph(G_web, weakly_components_sorted[40]) 
```

```python
nx.draw_circular(G40)
```

```python
#pagerank
def pagerank(G, iterations):
    
    #initialize pagerank scores for iteration 0
    scores = []
    for i in range(0, len(G)):
        scores.append(1/len(G))
    #max iterations 
    for i in range(0, iterations-1):
        #iterate through every node in graph
        for j in range(0, len(G)):
            node = list(G.nodes)[j]
            predecessors = list(G.predecessors(node))
            summ = 0
            #iterate through every predecessor
            for predecessor in predecessors:
                out_degree = G.out_degree(predecessor)
                summ += scores[j]/out_degree 
            #update pagerank score for iteration i
            scores[j] = summ
    #make a sorted list of tuples of final node pagerank scores
    result = []
    index = 0
    for node in G:
        element = (node, scores[index])
        result.append(element)
        index += 1
    return sorted(result, key = itemgetter(1), reverse=True)
```

```python
scores = pagerank(G40, len(G40))
print("Top 100 nodes according to pagerank algorithm:\n")
for score in scores[0:100]:
    print(score)
plt.figure()
plt.title("Pagerank Score Distribution")
plt.xlabel('score')
plt.ylabel('# nodes')
plt.hist([s for n, s in scores])
plt.show()
```

#### Δ1.1 Για κάθε έναν τύπο δικτύου, να παράξετε 100 διαφορετικές τοπολογίες και να παρουσιάσετε το διάγραμμα ποσοστού συνεκτικότητας σε σχέση με τις καθορισμένες παραμέτρους κάθε δικτύου.
#### Δ1.2 Σε ποιες περιπτώσεις εμφανίζονται φαινόμενα κατωφλίου;
#### Δ1.3 Ποιες είναι οι κρίσιμες τιμές για κάθε τύπο δικτύου οι οποίες οδηγούν σε μετάβαση φάσης (phase transitions);
#### Δ1.4 Ποιες μεταβάσεις φάσης είναι απότομες (sharp) και ποιες ομαλές (smooth);

```python
def REG_connectivity(n, d):
    connectivity = 0
    for i in range(100):
        tmp_graph = nx.watts_strogatz_graph(n, d, 0)
        connectivity += nx.is_connected(tmp_graph)
    return connectivity/100

plt.figure(1)
plt.title('REG n=100')
plt.xlabel('Degree')
plt.ylabel('Connectivity Probability')
X = np.arange(2,10+2,2)
Y = [REG_connectivity(100, d) for d in X]
plt.bar(X, Y, width=1)

plt.figure(2)
plt.title('REG n=200')
plt.xlabel('Degree')
Y = [REG_connectivity(200, d) for d in X]
plt.bar(X, Y, width=1)
```

### REG Connectivity

Λαμβάνουμε ποσοστό συνεκτικότητας 100%, αφού από τον ορισμό της τοπολογίας κάθε κόμβος συνδέεται πάντα με τους $k$ κοντινότερους γείτονές του. Οπότε κανένας δεν "μένει" εκτός δικτύου.

```python
def RGER_connectivity(n, M):
    connectivity = 0
    for i in range(100):
        tmp_graph = nx.gnm_random_graph(n, M)
        connectivity += nx.is_connected(tmp_graph)
    return connectivity/100

plt.figure(1)
plt.title('RGER n=100')
plt.xlabel('# Edges')
plt.ylabel('Connectivity Probability')
X = np.arange(100,800+100,100)
Y = [RGER_connectivity(100, M) for M in X]
plt.bar(X, Y, width=50)

plt.figure(2)
plt.title('RGER n=200')
plt.xlabel('# Edges')
plt.ylabel('Connectivity Probability')
Y = [RGER_connectivity(200, M) for M in X]
plt.bar(X, Y, width=50)
```

### RGER Connectivity

Δεν έχουμε φαινόμενο κατωφλίου. Παρατηρούμε για n=100 απότομη μετάβαση φάσης (κρίσιμες τιμές Μ = 100-400). Ενώ για n=200 παρατηρούμε ομαλή φάση μετάβασης με κρίσιμες τιμές του Μ από 400 έως 800.

```python
def RG_connectivity(n, p):
    connectivity = 0
    for i in range(100):
        tmp_graph = nx.random_geometric_graph(n, p) 
        connectivity += nx.is_connected(tmp_graph)
    return connectivity/100

plt.figure(1)
plt.title('RG-Gilbert n=100')
plt.xlabel('Probability')
plt.ylabel('Connectivity Probability')
X = np.arange(0.01,0.11,0.01)
Y = [RG_connectivity(100, p) for p in X]
plt.bar(X, Y, width=0.01)

plt.figure(2)
plt.title('RG-Gilbert n=200')
plt.xlabel('Probability')
plt.ylabel('Connectivity Probability')
Y = [RG_connectivity(200, p) for p in X]
plt.bar(X, Y, width=0.01)
```

### RG-Gilbert Connectivity

Για n=100 παρατηρούμε πως για τις δοσμένες τιμές R δεν υπάρχει καθόλου συνεκτικότητα.
Για n=200 παρατηρούμε πως πάλι δεν υπάρχει συνεκτικότητα, εκτώς από την τιμή R=0.10 όπου φαίνεται να αρχίζει μία μετάβαση φάσης. Οι τιμές του R είναι αρκετά μικρές και επομένως δεν μας αφήνουν να εξάγουμε συμπεράσματα για τις κρίσιμες τιμές ή τον τύπο μετάβασης φάσης (ομαλός ή απότομος). 

```python
def RGG_connectivity(n, R):
    connectivity = 0
    for i in range(100):
        tmp_graph = nx.random_geometric_graph(n, R)
        connectivity += nx.is_connected(tmp_graph)
    return connectivity/100

plt.figure(1)
plt.title('RGG n=100')
plt.xlabel('Radius')
plt.ylabel('Connectivity Probability')
X = np.arange(0.025,0.275,0.025)
Y = [RGG_connectivity(100, R) for R in X]
plt.bar(X, Y, width=0.0125)

plt.figure(2)
plt.title('RGG n=200')
plt.xlabel('Radius')
plt.ylabel('Connectivity Probability')
Y = [RGG_connectivity(200, R) for R in X]
plt.bar(X, Y, width=0.0125)
```

### RGG Connectivity

Δεν παρατηρούνται φαινόμενα κατωφλίου. 
Για n=100 έχουμε ομαλή μετάβαση φάσης με κρίσιμες τιμές R=0.15-0.20
Για n=200 έχουμε πάλι ομαλή προς απότομη μετάβαση φάσης με κρίσιμες τιμές 0.10-0.15

```python
def SF_connectivity(n, d):
    connectivity = 0
    for i in range(100):
        tmp_graph = nx.barabasi_albert_graph(n, d)
        connectivity += nx.is_connected(tmp_graph)
    return connectivity/100

plt.figure(1)
plt.title('SF n=100')
plt.xlabel('# Edges')
plt.ylabel('Connectivity Probability')
X = np.arange(2, 10+2, 2)
Y = [RGG_connectivity(100, d) for d in X]
plt.bar(X, Y, width=1)

plt.figure(2)
plt.title('SF n=200')
plt.xlabel('# Edges')
plt.ylabel('Connectivity Probability')
Y = [RGG_connectivity(200, d) for d in X]
plt.bar(X, Y, width=1)
```

### SF Connectivity

Ανεξάρτητα από την τιμή των παραμέτρων παρατηρούμε 100% συνεκτικότητα. Αυτό συμβαίνει καθώς από την ιδιότητα της τοπολογίας Scale Free, οι νέοι προστηθέμενοι κόμβοι συνδέονται με τους hub-κόμβους.

```python
def SW_connectivity(n, d, p):
    connectivity = 0
    for i in range(100):
        tmp_graph = nx.connected_watts_strogatz_graph(n, d, p)
        connectivity += nx.is_connected(tmp_graph)
    return connectivity/100

plt.figure(1)
plt.title('SW n=100')
plt.xlabel('d + 2*p')
plt.ylabel('Connectivity Probability')
X = [d+2*p for d in np.arange(2,10+2,2) for p in np.arange(0.1,0.7+0.1,0.1)]
Y = [SW_connectivity(100, d, p) for d in np.arange(2,10+2,2) for p in np.arange(0.1, 0.7+0.1,0.1)]
plt.bar(X, Y, width = 0.05)

plt.figure(2)
plt.title('SW n=200')
plt.xlabel('d + 2*p')
plt.ylabel('Connectivity Probability')
Y = [SW_connectivity(200, d, p) for d in np.arange(2,10+2,2) for p in np.arange(0.1, 0.7+0.1,0.1)]
plt.bar(X, Y, width = 0.05)
```

### SW Connectivity

Όμοια με SF τοπολογίες, λαμβάνουμε 100% συνδεσιμότητα που οφείλεται στην συνάρτηση που χρησιμοποιούμε που απαιτεί οι γράφοι να είναι συνδεδεμένοι εξαρχής.

### E) Μελέτη της εξελικτικής μετατροπής δικτύου REG σε δίκτυο SW και RGER

##### Μεταβάλλοντας την παράμετρο $g_p$ από $0$ έως $1$ με βήμα $0.1$της συνάρτησης που κατασκευάζει τοπολογίες SW μεγέθους $n=30x$ ( αντικαταστήστε το x με το τελευταίο ψηφίο του ΑΜ) με βαθμό αρχικού πλέγματος $d=4$, να βρείτε το μέσο μήκος μονοπατιού και το μέσο συντελεστή ομαδοποίησης. Πώς συμπεριφέρεται το παραγόμενο δίκτυο για $g_p = 0$, πώς για $g_p = 1$ και πώς για ενδιάμεσες τιμές $g_p \simeq 0.1 - 0.5$;

```python
random.seed(seedno)
n=307
d=4
average_shortest_path = []
mean_clust_coeff = []
X = np.arange(0, 1+0.1, 0.1)
for gp in X:
    G = nx.watts_strogatz_graph(n, d, gp)
    average_shortest_path.append(nx.average_shortest_path_length(G))
    mean_clust_coeff.append(np.mean(list(nx.clustering(G).values())))
plt.figure(1)
plt.title('Average Shortest Path in SW topologies')
plt.ylabel('Average Shortest Path')
plt.xlabel('gp')
plt.plot(X, average_shortest_path, marker='o')
plt.figure(2)
plt.title('Clustering Coefficient Mean')
plt.ylabel('Clustering Coefficient Mean')
plt.xlabel('gp')
plt.plot(X, mean_clust_coeff, marker='o')
```

### Σχολισαμός αποτελεσμάτων 

Για $gp=0$ το παραγόμενο δίκτυο φαίνεται να είναι τύπου RGG, αφού για αυτήν την τιμή παρουσιάζεται το μεγαλύτερο μέσο μικρότερο μονοπάτι καθώς και ο μεγαλύτερος συντελεστής ομαδοποίησης.
Για $gp=1$ το παραγόμενο δίκτυο φαίνεται να είναι τυχαίο, αφού τα μονοπάτια είναι πολύ μικρά αλλά και ο μέσος συντελεστής ομαδοποίησης.
Για ενδιάμεσες τιμές του $gp$ μεταξύ $0.1 - 0.5$ παρατηρούμε πως έχουμε μικρά μονοπάτια και μεσαίες τιμές μέσου συντελεστή ομαδοποίησης, κάτι που υποδηλώνει τοπολογία SW.
